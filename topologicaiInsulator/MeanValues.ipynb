```python
# Y mean
from numba import njit
from tqdm import tqdm
En = np.asarray(En, dtype=float)
Cn20 = np.asarray(Cn20, dtype=complex)
Cn30 = np.asarray(Cn30, dtype=complex)
Cn40 = np.asarray(Cn40, dtype=complex)
Cn50= np.asarray(Cn50, dtype=complex)
Cn80 = np.asarray(Cn80, dtype=complex)
Ynm = np.asarray(Ynm, dtype=complex)
@njit 
def meanvalue(t):
    w=(En[1]-En[0])
    T=2.*np.pi/w
    dt=T/20.
    Ymean20 = 0. + 1j*0.
    Ymean30 = 0. + 1j*0.
    Ymean40 = 0. + 1j*0.
    Ymean50 = 0. + 1j*0.
    Ymean80 = 0. + 1j*0.
    for m in range(len(En)):
        for n in range(len(En)):
            Ymean20=Ymean20+np.conj(Cn20[t][m])*Cn20[t][n]*Ynm[m][n]*np.exp(-1j*(En[n]-En[m])*t*dt)
            Ymean30=Ymean30+np.conj(Cn30[t][m])*Cn30[t][n]*Ynm[m][n]*np.exp(-1j*(En[n]-En[m])*t*dt)
            Ymean40=Ymean40+np.conj(Cn40[t][m])*Cn40[t][n]*Ynm[m][n]*np.exp(-1j*(En[n]-En[m])*t*dt)
            Ymean50=Ymean50+np.conj(Cn50[t][m])*Cn50[t][n]*Ynm[m][n]*np.exp(-1j*(En[n]-En[m])*t*dt)
            Ymean80=Ymean80+np.conj(Cn80[t][m])*Cn80[t][n]*Ynm[m][n]*np.exp(-1j*(En[n]-En[m])*t*dt)
    return [Ymean20.real, Ymean30.real, Ymean40.real, Ymean50.real, Ymean80.real]
Ymean20Full = []
Ymean30Full = []
Ymean40Full = []
Ymean50Full = []
Ymean80Full = []
tn=2490
for t in tqdm(range(tn)):
    MeanValues=meanvalue(t)
    Ymean20Full.append(MeanValues[0])
    Ymean30Full.append(MeanValues[1])
    Ymean40Full.append(MeanValues[2])
    Ymean50Full.append(MeanValues[3])
    Ymean80Full.append(MeanValues[4])
Ymean80Fullnew=[]
Ymean50Fullnew=[]
Ymean40Fullnew=[]
Ymean30Fullnew=[]
Ymean20Fullnew=[]
for i in range(tn):
    if ((i+1) % 20) == 0:
        Ymean80Fullnew.append(Ymean80Full[i])
        Ymean50Fullnew.append(Ymean50Full[i])
        Ymean40Fullnew.append(Ymean40Full[i])
        Ymean30Fullnew.append(Ymean30Full[i])
        Ymean20Fullnew.append(Ymean20Full[i])
```

```python
# Sx mean
fileSxRe = open("ReSx.dat")
dataSxRe=[]
for i in range(N):
    dataSxRe.append([])
    for j in range(N):
        dataSxRe[i].append(fileSxRe.readline())
fileSxRe.close()
Sxnm=[]
for i in range(N):
    Sxnm.append([])
    for j in range(N):
        Sxnm[i].append(float(dataSxRe[i][j]))
from numba import njit
from tqdm import tqdm
En = np.asarray(En, dtype=float)
Cn20 = np.asarray(Cn20, dtype=complex)
Cn30 = np.asarray(Cn30, dtype=complex)
Cn40 = np.asarray(Cn40, dtype=complex)
Cn50= np.asarray(Cn50, dtype=complex)
Cn80 = np.asarray(Cn80, dtype=complex)
Sxnm = np.asarray(Sxnm, dtype=float)
@njit 
def meanvalue(t):
    w=(En[1]-En[0])
    T=2.*np.pi/w
    dt=T/20.
    Sxmean20 = 0. + 1j*0.
    Sxmean30 = 0. + 1j*0.
    Sxmean40 = 0. + 1j*0.
    Sxmean50 = 0. + 1j*0.
    Sxmean80 = 0. + 1j*0.
    for m in range(len(En)):
        for n in range(len(En)):
            Sxmean20=Sxmean20+np.conj(Cn20[t][m])*Cn20[t][n]*Sxnm[m][n]*np.exp(-1j*(En[n]-En[m])*t*dt)
            Sxmean30=Sxmean30+np.conj(Cn30[t][m])*Cn30[t][n]*Sxnm[m][n]*np.exp(-1j*(En[n]-En[m])*t*dt)
            Sxmean40=Sxmean40+np.conj(Cn40[t][m])*Cn40[t][n]*Sxnm[m][n]*np.exp(-1j*(En[n]-En[m])*t*dt)
            Sxmean50=Sxmean50+np.conj(Cn50[t][m])*Cn50[t][n]*Sxnm[m][n]*np.exp(-1j*(En[n]-En[m])*t*dt)
            Sxmean80=Sxmean80+np.conj(Cn80[t][m])*Cn80[t][n]*Sxnm[m][n]*np.exp(-1j*(En[n]-En[m])*t*dt)
    return [Sxmean20.real, Sxmean30.real, Sxmean40.real, Sxmean50.real, Sxmean80.real]
Sxmean20Full = []
Sxmean30Full = []
Sxmean40Full = []
Sxmean50Full = []
Sxmean80Full = []
tn=2490
for t in tqdm(range(tn)):
    MeanValues=meanvalue(t)
    Sxmean20Full.append(MeanValues[0])
    Sxmean30Full.append(MeanValues[1])
    Sxmean40Full.append(MeanValues[2])
    Sxmean50Full.append(MeanValues[3])
    Sxmean80Full.append(MeanValues[4])
Sxmean80Fullnew=[]
Sxmean50Fullnew=[]
Sxmean40Fullnew=[]
Sxmean30Fullnew=[]
Sxmean20Fullnew=[]
for i in range(tn):
    if ((i+1) % 20) == 0:
        Sxmean80Fullnew.append(Sxmean80Full[i])
        Sxmean50Fullnew.append(Sxmean50Full[i])
        Sxmean40Fullnew.append(Sxmean40Full[i])
        Sxmean30Fullnew.append(Sxmean30Full[i])
        Sxmean20Fullnew.append(Sxmean20Full[i])
```
